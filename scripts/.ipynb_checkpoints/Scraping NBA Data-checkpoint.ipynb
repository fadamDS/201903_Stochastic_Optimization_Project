```python
initial_string = "www.basketball-reference.com"

##URLs to pull from:
initial_list = soup.select('#per_game_stats > tbody > tr > td > a')
gamelog_string_2018 = "/gamelog/2018"
gamelog_string_2019 = "/gamelog/2019"
http = 'https://'
```

```python
%%time

players_urls_2017_18 = []

for i in range (0, len(initial_list)):
    if (http + initial_string + initial_list[i]['href'][0:-5] + gamelog_string_2018) in players_urls_2017_18:
        continue
    else:
        if initial_list[i]['href'][1] == "p":
            players_urls_2017_18.append(http + initial_string + initial_list[i]['href'][0:-5] + gamelog_string_2018)
```

```python
%%time 
##Master loop

did_not_play = []
home_away = []
game_start = []
seconds_played = []
field_goal_pct = []
field_goal_attempt = []
field_goal_made = []
threePT_made = []
threePT_attempt = []
threePT_pct = []
free_throws_made = []
free_throws_attempt = []
free_throws_pct = []
orb = []
drb = []
assists = []
steals = []
blocks = []
turnovers = []
points = []
game_score = []
plus_minus = []

##1 = away
##0 = home


count = 0

for i in range(0, len(test_soup_grab_list)):
    
    ## if empty, they didn't play    
    if (souper_douper[i]['data-stat']) == 'game_season':
        if (souper_douper[i].text) == '':
            did_not_play.append(1)
            game_start.append(None)
            seconds_played.append(None)
            field_goal_made.append(None)
            field_goal_attempt.append(None)
            field_goal_pct.append(None)
            threePT_made.append(None)
            threePT_attempt.append(None)
            threePT_pct.append(None)
            free_throws_made.append(None)
            free_throws_attempt.append(None)
            free_throws_pct.append(None)
            orb.append(None)
            drb.append(None)
            assists.append(None)
            steals.append(None)
            blocks.append(None)
            turnovers.append(None)
            points.append(None)
            game_score.append(None)
            plus_minus.append(None)
            count = count + 1
        else:
            did_not_play.append(0)
            count = count + 1
    
    ##home or away
    if (souper_douper[i]['data-stat']) == 'game_location':
        if (souper_douper[i].text) == "@":
            home_away.append(1)
        else:
            home_away.append(0)
    
    ##games started        
    if (souper_douper[i]['data-stat']) == 'gs':   
        game_start.append(int(souper_douper[i].text))

    ##seconds played
    if (souper_douper[i]['data-stat']) == 'mp': 
        minute, sec = souper_douper[i].text.split(':')
        seconds_played.append(60*int(minute) + int(sec))      
        
    ##field goal made
    if (souper_douper[i]['data-stat']) == 'fg':
        field_goal_made.append(int(souper_douper[i].text))
    
    ##field goal attempted
    if (souper_douper[i]['data-stat']) == 'fga':
        field_goal_attempt.append(int(souper_douper[i].text))
        
    ##field goal percentage
    if (souper_douper[i]['data-stat']) == 'fg_pct':
        if (souper_douper[i].text) == '':
            field_goal_pct.append(None)
        else:
            field_goal_pct.append(float(souper_douper[i].text))
       
    ##3 pointers made
    if (souper_douper[i]['data-stat']) == 'fg3':
        threePT_made.append(int(souper_douper[i].text))
    
    ##3 pointers attempted
    if (souper_douper[i]['data-stat']) == 'fg3a':
        threePT_attempt.append(int(souper_douper[i].text))
        
    ##3 pointers percentage
    if (souper_douper[i]['data-stat']) == 'fg3_pct':
        if (souper_douper[i].text) == '':
            threePT_pct.append(None)
        else:
            threePT_pct.append(float(souper_douper[i].text))
    
    ##free throws made
    if (souper_douper[i]['data-stat']) == 'ft':
        free_throws_made.append(int(souper_douper[i].text))
            
    ##free throws attempted
    if (souper_douper[i]['data-stat']) == 'fta':
        free_throws_attempt.append(int(souper_douper[i].text))
        
    ##free throws percentage
    if (souper_douper[i]['data-stat']) == 'ft_pct':
        if (souper_douper[i].text) == '':
            free_throws_pct.append(None)
        else:
            free_throws_pct.append(float(souper_douper[i].text))

    ##offensive rebounds
    if (souper_douper[i]['data-stat']) == 'orb':
        orb.append(int(souper_douper[i].text))
        
    ##defensive rebounds
    if (souper_douper[i]['data-stat']) == 'drb':
        drb.append(int(souper_douper[i].text))
  
    ##assists
    if (souper_douper[i]['data-stat']) == 'ast':
        assists.append(int(souper_douper[i].text))
        
    ##steals
    if (souper_douper[i]['data-stat']) == 'stl':
        steals.append(int(souper_douper[i].text)) 
            
    ##blocks
    if (souper_douper[i]['data-stat']) == 'blk':
        blocks.append(int(souper_douper[i].text)) 
        
    ##turnovers
    if (souper_douper[i]['data-stat']) == 'tov':
        turnovers.append(int(souper_douper[i].text)) 
            
    ##points
    if (souper_douper[i]['data-stat']) == 'pts':
        points.append(int(souper_douper[i].text)) 
            
    ##game_score
    if (souper_douper[i]['data-stat']) == 'game_score':
        game_score.append(float(souper_douper[i].text))
            
    ##plus_minus
    if (souper_douper[i]['data-stat']) == 'plus_minus':
        if souper_douper[i].text == '':
            plus_minus.append(0)
        else:
            plus_minus.append(float(souper_douper[i].text))
```

```python
%%time 

Final_DF = pd.DataFrame(columns=['name',
                                 'did_not_play',
                                 'date',
                                 'team',
                                 'opponent',
                                 'home_away',
                                 'game_start',
                                 'seconds_played',
                                 'field_goal_pct',
                                 'field_goal_attempt',
                                 'field_goal_made',
                                 'threePT_made',
                                 'threePT_attempt',
                                 'threePT_pct',
                                 'free_throws_made',
                                 'free_throws_attempt',
                                 'free_throws_pct',
                                 'orb',
                                 'drb',
                                 'trb',
                                 'assists',
                                 'steals',
                                 'blocks',
                                 'turnovers',
                                 'points',
                                 'game_score',
                                 'plus_minus'
                                 ])

for j in range(0, len(players_urls_2017_18)):
    
    loop_response = requests.get(players_urls_2017_18[j])
    loop_soup = BeautifulSoup(loop_response.text, 'html.parser')
    
    souper_douper = loop_soup.select('#pgl_basic > tbody > tr > td')
    
    ##Master loop

    did_not_play = []
    home_away = []
    game_start = []
    seconds_played = []
    field_goal_pct = []
    field_goal_attempt = []
    field_goal_made = []
    threePT_made = []
    threePT_attempt = []
    threePT_pct = []
    free_throws_made = []
    free_throws_attempt = []
    free_throws_pct = []
    orb = []
    drb = []
    assists = []
    steals = []
    blocks = []
    turnovers = []
    points = []
    game_score = []
    plus_minus = []

    count = 0

    for i in range(0, len(souper_douper)):

        ## if empty, they didn't play    
        if (souper_douper[i]['data-stat']) == 'game_season':
            if (souper_douper[i].text) == '':
                did_not_play.append(1)
                game_start.append(None)
                seconds_played.append(None)
                field_goal_made.append(None)
                field_goal_attempt.append(None)
                field_goal_pct.append(None)
                threePT_made.append(None)
                threePT_attempt.append(None)
                threePT_pct.append(None)
                free_throws_made.append(None)
                free_throws_attempt.append(None)
                free_throws_pct.append(None)
                orb.append(None)
                drb.append(None)
                assists.append(None)
                steals.append(None)
                blocks.append(None)
                turnovers.append(None)
                points.append(None)
                game_score.append(None)
                plus_minus.append(None)
                count = count + 1
            else:
                did_not_play.append(0)
                count = count + 1

        ##home or away
        if (souper_douper[i]['data-stat']) == 'game_location':
            if (souper_douper[i].text) == "@":
                home_away.append(1)
            else:
                home_away.append(0)

        ##games started        
        if (souper_douper[i]['data-stat']) == 'gs':   
            game_start.append(int(souper_douper[i].text))

        ##seconds played
        if (souper_douper[i]['data-stat']) == 'mp': 
            minute, sec = souper_douper[i].text.split(':')
            seconds_played.append(60*int(minute) + int(sec))      

        ##field goal made
        if (souper_douper[i]['data-stat']) == 'fg':
            field_goal_made.append(int(souper_douper[i].text))

        ##field goal attempted
        if (souper_douper[i]['data-stat']) == 'fga':
            field_goal_attempt.append(int(souper_douper[i].text))

        ##field goal percentage
        if (souper_douper[i]['data-stat']) == 'fg_pct':
            if (souper_douper[i].text) == '':
                field_goal_pct.append(None)
            else:
                field_goal_pct.append(float(souper_douper[i].text))

        ##3 pointers made
        if (souper_douper[i]['data-stat']) == 'fg3':
            threePT_made.append(int(souper_douper[i].text))

        ##3 pointers attempted
        if (souper_douper[i]['data-stat']) == 'fg3a':
            threePT_attempt.append(int(souper_douper[i].text))

        ##3 pointers percentage
        if (souper_douper[i]['data-stat']) == 'fg3_pct':
            if (souper_douper[i].text) == '':
                threePT_pct.append(None)
            else:
                threePT_pct.append(float(souper_douper[i].text))

        ##free throws made
        if (souper_douper[i]['data-stat']) == 'ft':
            free_throws_made.append(int(souper_douper[i].text))

        ##free throws attempted
        if (souper_douper[i]['data-stat']) == 'fta':
            free_throws_attempt.append(int(souper_douper[i].text))

        ##free throws percentage
        if (souper_douper[i]['data-stat']) == 'ft_pct':
            if (souper_douper[i].text) == '':
                free_throws_pct.append(None)
            else:
                free_throws_pct.append(float(souper_douper[i].text))

        ##offensive rebounds
        if (souper_douper[i]['data-stat']) == 'orb':
            orb.append(int(souper_douper[i].text))

        ##defensive rebounds
        if (souper_douper[i]['data-stat']) == 'drb':
            drb.append(int(souper_douper[i].text))

        ##assists
        if (souper_douper[i]['data-stat']) == 'ast':
            assists.append(int(souper_douper[i].text))

        ##steals
        if (souper_douper[i]['data-stat']) == 'stl':
            steals.append(int(souper_douper[i].text)) 

        ##blocks
        if (souper_douper[i]['data-stat']) == 'blk':
            blocks.append(int(souper_douper[i].text)) 

        ##turnovers
        if (souper_douper[i]['data-stat']) == 'tov':
            turnovers.append(int(souper_douper[i].text)) 

        ##points
        if (souper_douper[i]['data-stat']) == 'pts':
            points.append(int(souper_douper[i].text)) 

        ##game_score
        if (souper_douper[i]['data-stat']) == 'game_score':
            game_score.append(float(souper_douper[i].text))

        ##plus_minus
        if (souper_douper[i]['data-stat']) == 'plus_minus':
            if souper_douper[i].text == '':
                plus_minus.append(0)
            else:
                plus_minus.append(float(souper_douper[i].text))
        
    ##after loop: make name, first 3 + trb
    name = []
    trb = []    
    date = []
    team = []
    opponent = []

    namesoup = loop_soup.select('#meta > div > h1')[0].text[0:-17]
    first3soup = loop_soup.select('#pgl_basic > tbody > tr > td > a')
    
    for i in range(0, len(did_not_play)):
        #name
        name.append(namesoup)
        
        #trb
        if orb[i] == None:
            trb.append(None)
        else:
            trb.append(orb[i] + drb[i])
      
    for i in range(0, len(first3soup)):
        if (i % 3) == 0:
            date.append(first3soup[i].text)
        elif (i % 3) == 1:
            team.append(first3soup[i].text)
        else:
            opponent.append(first3soup[i].text)
   
    intermediate_DF = pd.DataFrame(
                                    {'name': name,
                                     'did_not_play': did_not_play,
                                     'date': date,
                                     'team': team,
                                     'opponent': opponent,
                                     'home_away': home_away,
                                     'game_start': game_start,
                                     'seconds_played': seconds_played,
                                     'field_goal_pct': field_goal_pct,
                                     'field_goal_attempt': field_goal_attempt,
                                     'field_goal_made': field_goal_made,
                                     'threePT_made': threePT_made,
                                     'threePT_attempt': threePT_attempt,
                                     'threePT_pct': threePT_pct,
                                     'free_throws_made': free_throws_made,
                                     'free_throws_attempt': free_throws_attempt,
                                     'free_throws_pct': free_throws_pct,
                                     'orb': orb,
                                     'drb': drb,
                                     'trb': trb,
                                     'assists': assists,
                                     'steals': steals,
                                     'blocks': blocks,
                                     'turnovers': turnovers,
                                     'points': points,
                                     'game_score': game_score,
                                     'plus_minus': plus_minus
                                    })

    Final_DF = pd.concat([Final_DF,
                          intermediate_DF],
                          axis = 0, ignore_index=True)
```

```python
%%time 

Final_DF_18_19 = pd.DataFrame(columns=['name',
                                         'did_not_play',
                                         'date',
                                         'team',
                                         'opponent',
                                         'home_away',
                                         'game_start',
                                         'seconds_played',
                                         'field_goal_pct',
                                         'field_goal_attempt',
                                         'field_goal_made',
                                         'threePT_made',
                                         'threePT_attempt',
                                         'threePT_pct',
                                         'free_throws_made',
                                         'free_throws_attempt',
                                         'free_throws_pct',
                                         'orb',
                                         'drb',
                                         'trb',
                                         'assists',
                                         'steals',
                                         'blocks',
                                         'turnovers',
                                         'points',
                                         'game_score',
                                         'plus_minus'
                                         ])

for j in range(0, len(players_urls_2018_19)):
    
    loop_response = requests.get(players_urls_2018_19[j])
    loop_soup = BeautifulSoup(loop_response.text, 'html.parser')
    
    souper_douper = loop_soup.select('#pgl_basic > tbody > tr > td')
    
    ##Master loop

    did_not_play = []
    home_away = []
    game_start = []
    seconds_played = []
    field_goal_pct = []
    field_goal_attempt = []
    field_goal_made = []
    threePT_made = []
    threePT_attempt = []
    threePT_pct = []
    free_throws_made = []
    free_throws_attempt = []
    free_throws_pct = []
    orb = []
    drb = []
    assists = []
    steals = []
    blocks = []
    turnovers = []
    points = []
    game_score = []
    plus_minus = []

    count = 0

    for i in range(0, len(souper_douper)):

        ## if empty, they didn't play    
        if (souper_douper[i]['data-stat']) == 'game_season':
            if (souper_douper[i].text) == '':
                did_not_play.append(1)
                game_start.append(None)
                seconds_played.append(None)
                field_goal_made.append(None)
                field_goal_attempt.append(None)
                field_goal_pct.append(None)
                threePT_made.append(None)
                threePT_attempt.append(None)
                threePT_pct.append(None)
                free_throws_made.append(None)
                free_throws_attempt.append(None)
                free_throws_pct.append(None)
                orb.append(None)
                drb.append(None)
                assists.append(None)
                steals.append(None)
                blocks.append(None)
                turnovers.append(None)
                points.append(None)
                game_score.append(None)
                plus_minus.append(None)
                count = count + 1
            else:
                did_not_play.append(0)
                count = count + 1

        ##home or away
        if (souper_douper[i]['data-stat']) == 'game_location':
            if (souper_douper[i].text) == "@":
                home_away.append(1)
            else:
                home_away.append(0)

        ##games started        
        if (souper_douper[i]['data-stat']) == 'gs':   
            game_start.append(int(souper_douper[i].text))

        ##seconds played
        if (souper_douper[i]['data-stat']) == 'mp': 
            minute, sec = souper_douper[i].text.split(':')
            seconds_played.append(60*int(minute) + int(sec))      

        ##field goal made
        if (souper_douper[i]['data-stat']) == 'fg':
            field_goal_made.append(int(souper_douper[i].text))

        ##field goal attempted
        if (souper_douper[i]['data-stat']) == 'fga':
            field_goal_attempt.append(int(souper_douper[i].text))

        ##field goal percentage
        if (souper_douper[i]['data-stat']) == 'fg_pct':
            if (souper_douper[i].text) == '':
                field_goal_pct.append(None)
            else:
                field_goal_pct.append(float(souper_douper[i].text))

        ##3 pointers made
        if (souper_douper[i]['data-stat']) == 'fg3':
            threePT_made.append(int(souper_douper[i].text))

        ##3 pointers attempted
        if (souper_douper[i]['data-stat']) == 'fg3a':
            threePT_attempt.append(int(souper_douper[i].text))

        ##3 pointers percentage
        if (souper_douper[i]['data-stat']) == 'fg3_pct':
            if (souper_douper[i].text) == '':
                threePT_pct.append(None)
            else:
                threePT_pct.append(float(souper_douper[i].text))

        ##free throws made
        if (souper_douper[i]['data-stat']) == 'ft':
            free_throws_made.append(int(souper_douper[i].text))

        ##free throws attempted
        if (souper_douper[i]['data-stat']) == 'fta':
            free_throws_attempt.append(int(souper_douper[i].text))

        ##free throws percentage
        if (souper_douper[i]['data-stat']) == 'ft_pct':
            if (souper_douper[i].text) == '':
                free_throws_pct.append(None)
            else:
                free_throws_pct.append(float(souper_douper[i].text))

        ##offensive rebounds
        if (souper_douper[i]['data-stat']) == 'orb':
            orb.append(int(souper_douper[i].text))

        ##defensive rebounds
        if (souper_douper[i]['data-stat']) == 'drb':
            drb.append(int(souper_douper[i].text))

        ##assists
        if (souper_douper[i]['data-stat']) == 'ast':
            assists.append(int(souper_douper[i].text))

        ##steals
        if (souper_douper[i]['data-stat']) == 'stl':
            steals.append(int(souper_douper[i].text)) 

        ##blocks
        if (souper_douper[i]['data-stat']) == 'blk':
            blocks.append(int(souper_douper[i].text)) 

        ##turnovers
        if (souper_douper[i]['data-stat']) == 'tov':
            turnovers.append(int(souper_douper[i].text)) 

        ##points
        if (souper_douper[i]['data-stat']) == 'pts':
            points.append(int(souper_douper[i].text)) 

        ##game_score
        if (souper_douper[i]['data-stat']) == 'game_score':
            game_score.append(float(souper_douper[i].text))

        ##plus_minus
        if (souper_douper[i]['data-stat']) == 'plus_minus':
            if souper_douper[i].text == '':
                plus_minus.append(0)
            else:
                plus_minus.append(float(souper_douper[i].text))
        
    ##after loop: make name, first 3 + trb
    name = []
    trb = []    
    date = []
    team = []
    opponent = []

    namesoup = loop_soup.select('#meta > div > h1')[0].text[0:-17]
    first3soup = loop_soup.select('#pgl_basic > tbody > tr > td > a')
    
    for i in range(0, len(did_not_play)):
        #name
        name.append(namesoup)
        
        #trb
        if orb[i] == None:
            trb.append(None)
        else:
            trb.append(orb[i] + drb[i])
      
    for i in range(0, len(first3soup)):
        if (i % 3) == 0:
            date.append(first3soup[i].text)
        elif (i % 3) == 1:
            team.append(first3soup[i].text)
        else:
            opponent.append(first3soup[i].text)
   
    intermediate_DF = pd.DataFrame(
                                    {'name': name,
                                     'did_not_play': did_not_play,
                                     'date': date,
                                     'team': team,
                                     'opponent': opponent,
                                     'home_away': home_away,
                                     'game_start': game_start,
                                     'seconds_played': seconds_played,
                                     'field_goal_pct': field_goal_pct,
                                     'field_goal_attempt': field_goal_attempt,
                                     'field_goal_made': field_goal_made,
                                     'threePT_made': threePT_made,
                                     'threePT_attempt': threePT_attempt,
                                     'threePT_pct': threePT_pct,
                                     'free_throws_made': free_throws_made,
                                     'free_throws_attempt': free_throws_attempt,
                                     'free_throws_pct': free_throws_pct,
                                     'orb': orb,
                                     'drb': drb,
                                     'trb': trb,
                                     'assists': assists,
                                     'steals': steals,
                                     'blocks': blocks,
                                     'turnovers': turnovers,
                                     'points': points,
                                     'game_score': game_score,
                                     'plus_minus': plus_minus
                                    })

    Final_DF_18_19 = pd.concat([Final_DF_18_19,
                                intermediate_DF],
                                axis = 0, ignore_index=True)
```
